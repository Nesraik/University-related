```python
from scipy.stats import pearsonr
copydata = data.copy()
copydata = copydata.drop(columns=['gdp_per_capita'])

columns = list(copydata.columns)
```

```python
selected_feature = []
for feature in columns:
    correlation_coefficient,p_value = pearsonr(copydata[feature],data['gdp_per_capita'])
    if abs(correlation_coefficient) > 0.1 and p_value < 0.05:
        selected_feature.append(feature)
```

```python
selected_feature
```

```python
selected_feature.append('gdp_per_capita')
data = data[selected_feature]
```

```python
data.shape
```

Standardize

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data = scaler.fit_transform(data)
data = pd.DataFrame(data,columns=selected_feature)
```

```python
data.head()
```

Multivariate Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

x = data.drop(columns=['gdp_per_capita'])
y = data['gdp_per_capita']

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.3)
LR = LinearRegression()

LR.fit(x_train,y_train)

LR_predict = LR.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error,mean_squared_error

print(r2_score(y_test,LR_predict))
print(mean_absolute_error(y_test,LR_predict))
print(mean_squared_error(y_test,LR_predict))
print(root_mean_squared_error(y_test,LR_predict))
```
